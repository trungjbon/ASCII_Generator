```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("input.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

```python
CHAR_LIST = "$@B%8&WM#*oahkbdpqwmZO0QLCJUYXzcvunxrjft/\\|()1{}[]?-_+~<>i!lI;:,\"^`'. "

num_cols = 200
height, width = img.shape
cell_width = width / num_cols
cell_height = cell_width * 2
num_row = int(height / cell_height)

output_file = open("image.txt", "w")

for i in range(num_row):
    for j in range(num_cols):
        sub_img = img[int(i * cell_height): int((i + 1) * cell_height), 
                      int(j * cell_width): int((j + 1) * cell_width)]
        
        idx = int(np.mean(sub_img) / 255 * len(CHAR_LIST))
        output_file.write(CHAR_LIST[idx])
    output_file.write("\n")

output_file.close()
```
